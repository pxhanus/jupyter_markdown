```python
import json
import urllib.request
```

```python
with urllib.request.urlopen('http://spec.commonmark.org/0.28/spec.json') as data:
    web_spec = json.loads(data.read())
web_spec
```

```python
with open("spec.json") as data:
    local_spec = json.load(data)
local_spec
```
